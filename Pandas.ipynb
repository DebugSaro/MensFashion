```python
import pandas as pd
```

```python
df = pd.read_csv("/content/Pandas_practice1.csv")
```

```python
df
```

```python
df.info()
```

```python
df_clean = df.drop_duplicates()
```

```python
df_clean.isnull().sum()
```

```python
df_clean.dropna(inplace=True)
```

```python
df_clean['Unit Price'] = df_clean['Unit Price'].astype(int)
```

```python
df_clean['Unit Price'] = df_clean['Unit Price'].fillna(df_clean['Unit Price'].mean())
```

```python
df_clean
```

```python
df_clean.groupby('Product')['Total Revenue'].sum()
```

```python
df_clean.groupby('Product')['Units Sold'].sum().sort_values(ascending = False)
```
